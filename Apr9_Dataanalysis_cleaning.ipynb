```python
def find_non_numeric_rows(df, columns_to_check):
    """
    Checks for rows containing non-numeric data in specified columns of a DataFrame.

    Args:
        df (pd.DataFrame): The input DataFrame.
        columns_to_check (list): A list of column names to check for numeric data.

    Returns:
        pd.DataFrame: A DataFrame containing rows where non-numeric data was found
                      in the specified columns.
    """
    non_numeric_rows = pd.DataFrame()
    for col in columns_to_check:
        # Create a boolean mask for non-numeric values in the current column
        is_non_numeric = pd.to_numeric(df[col], errors='coerce').isna()

        # Filter the DataFrame for rows where the current column has a non-numeric value
        non_numeric_in_col = df[is_non_numeric]

        if not non_numeric_in_col.empty:
            print(f"\nNon-numeric values found in column '{col}':")
            print(non_numeric_in_col)
            non_numeric_rows = pd.concat([non_numeric_rows, non_numeric_in_col], ignore_index=True)

    if non_numeric_rows.empty:
        print("\nNo non-numeric values found in the specified columns.")
    else:
        print("\n--- Summary of Rows with Non-Numeric Data ---")
        print(non_numeric_rows)

    return non_numeric_rows

# Specify the columns to check for numerical data
columns_to_inspect = ['DOCUMENT NO', 'QUANTITY IN', 'QUANTITY OUT', 'BALANCE']

# Apply the function to your df1 DataFrame
non_numeric_df = find_non_numeric_rows(df1.copy(), columns_to_inspect)
```

```python
def drop_rows_with_non_numeric(df, column_to_check):
    """
    Drops rows from a DataFrame where a specified column contains non-numeric data.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_to_check (str): The name of the column to check for numeric data.

    Returns:
        pd.DataFrame: A new DataFrame with rows containing non-numeric data
                      in the specified column dropped.
        pd.DataFrame: A DataFrame containing the dropped rows.
    """
    # Create a boolean mask for non-numeric values in the specified column
    is_non_numeric = pd.to_numeric(df[column_to_check], errors='coerce').isna()

    # Filter the DataFrame to get the rows with non-numeric values
    rows_to_drop = df[is_non_numeric].copy()

    # Drop the rows with non-numeric values from the original DataFrame
    df_cleaned = df[~is_non_numeric].copy()

    print(f"Shape of DataFrame before dropping rows: {df.shape}")
    print(f"Number of rows dropped (non-numeric in '{column_to_check}'): {len(rows_to_drop)}")
    print(f"Shape of DataFrame after dropping rows: {df_cleaned.shape}")

    if not rows_to_drop.empty:
        print("\n--- Rows dropped due to non-numeric values in column '{column_to_check}' ---")
        print(rows_to_drop)
    else:
        print(f"\nNo non-numeric values found in column '{column_to_check}'. No rows dropped.")

    return df_cleaned, rows_to_drop

# Specify the column to check for non-numeric data and drop rows accordingly
column_to_clean = 'QUANTITY OUT'

# Apply the function to your df1 DataFrame
df1_cleaned, dropped_rows_qty_out = drop_rows_with_non_numeric(df1.copy(), column_to_clean)

# Now 'df1_cleaned' contains the DataFrame with rows where 'QUANTITY OUT'
# had non-numeric values removed. 'dropped_rows_qty_out' contains the rows that were dropped.

# You can now work with 'df1_cleaned' for further analysis.
# If you want to see the cleaned DataFrame:
# print("\n--- Cleaned DataFrame (df1_cleaned) ---")
# print(df1_cleaned.head())
# print(df1_cleaned.tail())
# print(df1_cleaned.info())
```

```python
def check_duplicate_values(df, column_name):
    """
    Checks for duplicate values in a specified column of a DataFrame and counts their occurrences.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the column to check for duplicates.

    Returns:
        pandas.Series: A Series where the index is the duplicate value and the values
                       are the number of times that value appears (count > 1).
                       Returns an empty Series if no duplicates are found.
    """
    if column_name not in df.columns:
        print(f"Error: Column '{column_name}' not found in the DataFrame.")
        return pd.Series()

    duplicate_counts = df[column_name].value_counts()
    duplicate_values = duplicate_counts[duplicate_counts > 1]

    if not duplicate_values.empty:
        print(f"Duplicate values found in column '{column_name}':")
        print(duplicate_values)
    else:
        print(f"No duplicate values found in column '{column_name}'.")

    return duplicate_values

# Assuming your DataFrame is named 'df1_cleaned' after the previous step
column_to_check = 'DOCUMENT NO'  # Specify the column name

# Check for duplicates in the 'DOCUMENT NO' column
duplicate_series = check_duplicate_values(df1_cleaned, column_to_check)

# If you want to see the rows containing duplicates:
if not duplicate_series.empty:
    duplicate_rows_df = df1_cleaned[df1_cleaned['DOCUMENT NO'].isin(duplicate_series.index)]
    print("\n--- Rows containing duplicate values in the 'DOCUMENT NO' column ---")
    print(duplicate_rows_df)
```

```python
def process_dataframe(df):
    """
    Makes the 0-indexed row the header and deletes the last two rows.

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        pd.DataFrame: The processed DataFrame.
    """
    if len(df) > 0:
        # Set the first row as the new header
        df.columns = df.iloc[0]
        df = df[1:]  # Remove the original header row

        # Delete the last two rows if the DataFrame has at least two rows
        if len(df) >= 2:
            df = df[:-2]
        elif len(df) == 1:
            df = df[:-1]  # Delete the last one row
        # If DataFrame has 0 rows, nothing to delete

    return df

# Apply the processing function to your df2 DataFrame
df2_processed = process_dataframe(df2.copy())

# Now df2_processed contains the DataFrame with the 0-indexed row as header
# and the last two rows removed.

# You can inspect the processed DataFrame:
print("--- Processed DataFrame (df2_processed) ---")
print(df2_processed.head())
print(df2_processed.tail())
print(df2_processed.info())
```

```python
def delete_empty_rows_columns(df):
    """
    Deletes rows and columns that are entirely empty (contain only NaN values).

    Args:
        df (pd.DataFrame): The input DataFrame.

    Returns:
        pd.DataFrame: The DataFrame with empty rows and columns removed.
    """
    print(f"Shape before removing empty rows/columns: {df.shape}")

    # Delete empty rows (all values in the row are NaN)
    df_no_empty_rows = df.dropna(how='all')
    print(f"Shape after removing empty rows: {df_no_empty_rows.shape}")

    # Delete empty columns (all values in the column are NaN)
    df_no_empty_rc = df_no_empty_rows.dropna(axis=1, how='all')
    print(f"Shape after removing empty columns: {df_no_empty_rc.shape}")

    return df_no_empty_rc

# Assuming your DataFrame after setting the header and removing last two rows is 'df2_processed'
df2_cleaned = delete_empty_rows_columns(df2_processed.copy())

# Now df2_cleaned contains the DataFrame with empty rows and columns removed.

# You can inspect the cleaned DataFrame:
print("\n--- Cleaned DataFrame (df2_cleaned) ---")
print(df2_cleaned.head())
print(df2_cleaned.tail())
print(df2_cleaned.info())
```

```python
import pandas as pd

def calculate_pieces_refined_v4(df):
    """
    Calculates 'Pieces', correctly handling commas in 'Qty' and preserving 'Rate' decimals initially.
    """
    print("--- Before Cleaning and Calculation ---")
    print(f"Data type of 'Gross': {df['Gross'].dtype}")
    print(f"Data type of 'Rate': {df['Rate'].dtype}")
    print(f"Data type of 'Qty': {df['Qty'].dtype}")

    # Clean 'Gross' column
    df['Gross'] = df['Gross'].astype(str).str.replace(',', '', regex=False)
    df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce').astype('Int64')

    # Clean 'Qty' column: remove commas and convert to numeric (float initially)
    df['Qty'] = df['Qty'].astype(str).str.strip().str.replace(',', '', regex=False)
    df['Qty'] = pd.to_numeric(df['Qty'], errors='coerce')

    # Clean 'Rate' column: remove commas and convert to numeric (float)
    df['Rate'] = df['Rate'].astype(str).str.strip().str.replace(',', '', regex=False)
    df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')

    print("\n--- After Initial Numeric Conversion (Float for Rate & Qty) ---")
    print(f"Data type of 'Gross': {df['Gross'].dtype}")
    print(f"Data type of 'Rate': {df['Rate'].dtype}")
    print(f"Data type of 'Qty': {df['Qty'].dtype}")
    print(f"NaN count in 'Gross': {df['Gross'].isna().sum()}")
    print(f"NaN count in 'Rate': {df['Rate'].isna().sum()}")
    print(f"NaN count in 'Qty': {df['Qty'].isna().sum()}")

    # Fill NaN with 0
    df['Rate'] = df['Rate'].fillna(0)
    df['Qty'] = df['Qty'].fillna(0)

    # Calculate 'Pieces' as float first to maintain precision
    df['Pieces'] = df.apply(lambda row: row['Gross'] / row['Rate'] / row['Qty']
                                      if row['Rate'] != 0 and row['Qty'] != 0 and pd.notna(row['Gross'])
                                      else pd.NA, axis=1)

    # Convert 'Qty' and 'Rate' to Integer after calculation (if needed for final format)
    df['Qty'] = df['Qty'].round().astype('Int64')
    df['Rate'] = df['Rate'].round().astype('Int64')
    df['Pieces'] = df['Pieces'].round().astype('Int64')

    print("\n--- After 'Pieces' Calculation and Final Integer Conversion ---")
    print(f"Data type of 'Qty': {df['Qty'].dtype}")
    print(f"Data type of 'Rate': {df['Rate'].dtype}")
    print(f"Data type of 'Pieces': {df['Pieces'].dtype}")
    print(f"NaN count in 'Pieces': {df['Pieces'].isna().sum()}")

    return df

# Assuming your DataFrame is 'df2_cleaned'
df2_with_pieces_refined_v4 = calculate_pieces_refined_v4(df2_cleaned.copy())

# Inspect the result
print("\n--- DataFrame with 'Pieces' Column (Refined v4) ---")
print(df2_with_pieces_refined_v4.head())
print(df2_with_pieces_refined_v4.tail())
print(df2_with_pieces_refined_v4.info())
```

```python
def calculate_pieces_as_int_fixed_v2(df):
    """
    Calculates 'Pieces' as integer, handling NaN before integer casting.
    """
    print("--- Before Cleaning and Calculation ---")
    print(f"Data type of 'Gross': {df['Gross'].dtype}")
    print(f"Data type of 'Rate': {df['Rate'].dtype}")
    print(f"Data type of 'Qty': {df['Qty'].dtype}")

    # Clean and convert to numeric, handling errors
    df['Gross'] = df['Gross'].astype(str).str.replace(',', '', regex=False)
    df['Gross'] = pd.to_numeric(df['Gross'], errors='coerce')

    df['Rate'] = pd.to_numeric(df['Rate'], errors='coerce')
    df['Qty'] = pd.to_numeric(df['Qty'], errors='coerce')

    print("\n--- After Numeric Conversion ---")
    print(f"NaN count in 'Gross': {df['Gross'].isna().sum()}")
    print(f"NaN count in 'Rate': {df['Rate'].isna().sum()}")
    print(f"NaN count in 'Qty': {df['Qty'].isna().sum()}")

    # Fill NaN values with 0 before rounding and casting to integer
    df['Gross_int'] = df['Gross'].fillna(0).round().astype('Int64')
    df['Rate_int'] = df['Rate'].fillna(0).round().astype('Int64')
    df['Qty_int'] = df['Qty'].fillna(0).round().astype('Int64')

    print("\n--- After Filling NaN and Rounding for Integer Conversion ---")
    print(f"NaN count in 'Gross_int': {df['Gross_int'].isna().sum()}")
    print(f"NaN count in 'Rate_int': {df['Rate_int'].isna().sum()}")
    print(f"NaN count in 'Qty_int': {df['Qty_int'].isna().sum()}")
    print(f"Data type of 'Gross_int': {df['Gross_int'].dtype}")
    print(f"Data type of 'Rate_int': {df['Rate_int'].dtype}")
    print(f"Data type of 'Qty_int': {df['Qty_int'].dtype}")

    # Calculate 'Pieces' as integer, avoiding division by zero
    df['Pieces'] = df.apply(lambda row: int(round(row['Gross_int'] / row['Rate_int'] / row['Qty_int']))
                                      if row['Rate_int'] != 0 and row['Qty_int'] != 0
                                      else pd.NA, axis=1).astype('Int64')

    print("\n--- After 'Pieces' Calculation (Integer) ---")
    print(f"NaN count in 'Pieces': {df['Pieces'].isna().sum()}")
    print(f"Data type of 'Pieces': {df['Pieces'].dtype}")

    # Drop the intermediate integer columns
    df = df.drop(columns=['Gross_int', 'Rate_int', 'Qty_int'])

    return df

# Assuming your DataFrame is 'df2_cleaned'
df2_with_pieces_int_fixed_v2 = calculate_pieces_as_int_fixed_v2(df2_cleaned.copy())

# You can now inspect the DataFrame:
print("\n--- DataFrame with 'Pieces' Column (Integer Fixed v2) ---")
print(df2_with_pieces_int_fixed_v2.head())
print(df2_with_pieces_int_fixed_v2.tail())
print(df2_with_pieces_int_fixed_v2.info())
```
